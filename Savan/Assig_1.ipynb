```python
import cv2

img = cv2.imread(r'C:\Users\allen\Desktop\OpenCv\8.png')
cv2.imshow('shapes', img)
cv2.waitKey(0)
cv2.destroyAllWindows() 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

ret, thresh = cv2.threshold(gray,127,255,1)

contours, hierarchy = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
image=img
cv2.drawContours(img, contours, -1, (0, 255, 255), 5)
cv2.imshow('shapes', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python


length=[]
for cnt in contours:
  
    approx = cv2.approxPolyDP(cnt, 0.01 * cv2.arcLength(cnt, True), True)
    length.append(len(approx))
length.sort()
length

M = cv2.moments(cnt)
if M['m00'] != 0.0:
    x = int(M['m10']/M['m00'])
    y = int(M['m01']/M['m00'])     
    
image = cv2.imread(r'C:\Users\allen\Desktop\OpenCv\8.png')
img = image
for i in range(0,5):
    
    for cnt in contours:
        
        approx = cv2.approxPolyDP(cnt,0.01*cv2.arcLength(cnt,True),True)
        if len(approx) == length[i]:
            if len(approx) == 3:
                cv2.drawContours(image,[cnt],0,(0,0,255),-1)
                M = cv2.moments(cnt)
                if M['m00'] != 0.0:
                    x = int(M['m10']/M['m00'])
                    y = int(M['m01']/M['m00'])
                cv2.putText(img, 'No.1 Triangle=3', (x, y),cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1)
                
            elif len(approx) == 4:
                cv2.drawContours(image,[cnt],0,(0,255,0),-1)
                M = cv2.moments(cnt)
                if M['m00'] != 0.0:
                    x = int(M['m10']/M['m00'])
                    y = int(M['m01']/M['m00'])
                cv2.putText(img, 'No.2 Quadrilateral=4', (x, y),cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1)
                
            elif len(approx) == 5:
                cv2.drawContours(image,[cnt],0,(255,0,0),-1)
                M = cv2.moments(cnt)
                if M['m00'] != 0.0:
                    x = int(M['m10']/M['m00'])
                    y = int(M['m01']/M['m00'])
                cv2.putText(img, 'No.3 Pentagon=5', (x, y),cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1)
                
            elif len(approx) == 6:
                cv2.drawContours(image,[cnt],0,(255,255,0),-1)
                M = cv2.moments(cnt)
                if M['m00'] != 0.0:
                    x = int(M['m10']/M['m00'])
                    y = int(M['m01']/M['m00'])
                cv2.putText(img, 'No.4 Hexagon=6', (x, y),cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1)
                
            else:
                cv2.drawContours(image,[cnt],0,(0,255,255),-1)
                M = cv2.moments(cnt)
                if M['m00'] != 0.0:
                    x = int(M['m10']/M['m00'])
                    y = int(M['m01']/M['m00'])
                cv2.putText(img, 'No.5 circle', (x, y),cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1)
                
            
            cv2.imshow('identified shapes',image)
            cv2.waitKey(0)
                        
cv2.waitKey(0)
cv2.destroyAllWindows()
```
